# Visualizing Historical Trends

Data often have a sequence. Economic indicators are reported every month. In the
United States, population estimates are gathered every ten years.  Social
network events are timestamped with second or millisecond precision. Printed
texts also frequently come in a sequence. Newspapers and periodicals have
volumes.  Novels have chapters. Personal diaries have dated entries. Taking into
consideration the temporal sequence of data often reveals surprising patterns.

## British Novels, 1800--1829

As a motivating example, consider the number of novels published in the British
Isles between 1800 and 1829.^[These data come from [@garside2000].] For each
year we have the number of novels published by men, women, and authors of
unknown gender. (Publishing anonymously was a common practice in the 18th and
19th century.)

````python
import pandas as pd
novels = pd.read_csv('data/novels-1800-1829.csv', index_col=0)
novels.head(10)
````

There are a number of ways these data might be visualized. The most familiar
method for visualizing a time series when data is evenly spaced and there are no
missing values is a plot which creates a line connecting each data point.
Invariably the year will be placed on the horizontal axis (or x-axis) and the
relevant number on the vertical axis (or y-axis). In this case, the number of
novels published in each year is indicated on the vertical axis.


````python
novels.plot(title="British Novels, 1800-1829")
````

This data set concerns a period before the emergence of a "mass market" for
novels and the number of novels published each year is correspondingly modest by
contemporary standards. In order to look for an overall trend, we merge the
number of novels published in the same year for each category of
author:^[Specifically, we sum across the columns of the table. Here the Pandas
library adopts NumPy's convention of referring to the traversing of the columns
"axis" as moving across axis 1. The axis corresponding to the rows is axis 0.]


````python
novels.sum(axis=1).plot(title="British Novels, 1800-1829")
````

Looking at the first plot, we can see that the demographic makeup of novel
writers is changing between 1800 and 1829. One way to make these changes more
visible is to group the publication figures by decade. In order to do this, we
first need to add a variable indicating to which decade a particular observation
belongs. A quick way to get the decade from an integer year is to divide by the
number ten and report only the integer quotient. Python happens to have an
operator, ``//`` that performs precisely this operation ("integer division"):


````python
novels['decade'] = (novels.index // 10) * 10
novels.head(12)
````

Now that we have the decade associated with each record of publication counts,
we must group and sum counts linked with the same decade. This operation is
known as a "group by" operation in a variety of settings.^[In SQL, for example,
the relevant command is "GROUP BY".] When dealing with a Pandas ``DataFrame``,
the method's name is ``groupby``. ``groupby`` yields an intermediate abstract
object upon which we are expected to perform some subsequent operation. In this
case, the operation is summing, for which we use the method ``sum``:


````python
novels.groupby('decade').sum().head()
````

Because our data is no longer evenly spaced, drawing a line to connect the
counts becomes less appealing as a visualization strategy. In this context,
a bar chart grouped by decade is better suited:


````python
novels.groupby('decade').sum().plot(kind='bar', title="British Novels, 1800-1829")
````

### Converting among different formats

Too often the data we have is not in the format we need. The same data can be
represented in a wide variety of formats. Converting from one representation to
another is a common chore in data analysis. For example, consider what steps
would be required if we were given the data used in the examples above in the
following disaggregated format, where each row is an observation and each column
is a variable:


````python
novels_alt = pd.read_csv('data/novels-1800-1829-alt.csv') 
novels_alt.head()                                    
````

(Those preferring this format have dubbed it ["tidy" data](http://vita.had.co.nz/papers/tidy-data.pdf).)
<!-- CITE: @Article{tidy-data, author = {Hadley Wickham}, journal = {The Journal of Statistical Software}, title = {Tidy data}, year = {Submitted}, } -->

Converting this format into the format used earlier requires "pivoting" the data
such that the desired variables are associated with the rows and columns. In
this case, we want the ``gender`` variable to occupy the columns and the
``year`` variable to occupy the rows. The interior values of the dataset will be
taken from the ``count`` variable:


````python
novels_pivot = novels_alt.pivot('year', 'gender', 'count')
novels_pivot.head()
````

## Les Misérables

Another example of a sequence we might visualize concerns the unfolding of events in Victor Hugo's *Les Misérables*. Over 500,000
words long, the book counts as a lengthy text by any standard.[^1] The novel
comes in five volumes ("Fantine", "Cosette", "Marius", "The Idyll in the Rue
Plumet and the Epic in the Rue St.  Denis", and "Jean Valjean"). Within each
volume we have a sequence of chapters and within each chapter we have a sequence
of paragraphs. In this chapter we address the visualization and processing of
sequential data.

Consider the rise and fall of a topic associated with revolutionary activity in
*Les Misérables* (Figure 1). This topic features words such as *barricade*
(barricade), *fusil* (rifle), and the character
[Enjolras](https://en.wikipedia.org/wiki/Enjolras), the leader of the
revolutionary *Les Amis de l'ABC*. This topic is derived from a topic model of
the novel (topic models are described in chapter~?).


````python
from IPython.display import Image
Image(filename="images/plot_topics_over_time_series_les_misérables.png")
````

As always, we first need to fit a topic model to the corpus. As MALLET has no
built-in French stopword list we need to provide one. We will use the [French
stopword
list](http://svn.tartarus.org/snowball/trunk/website/algorithms/french/stop.txt)
from the Snowball stemmer package. Additionally, because we are dealing with
non-English text we need to use an alternate regular expression for
tokenization. MALLET helpfully suggests `--token-regex '[\p{L}\p{M}]+'`.



```
mallet-2.0.7/bin/mallet import-dir --input data/hugo-les-misérables-split/ --output /tmp/topic-input-hugo.mallet --keep-sequence --remove-stopwords --stoplist-file data/stopwords/french.txt --token-regex '[\p{L}\p{M}]+'
mallet-2.0.7/bin/mallet train-topics --input /tmp/topic-input-hugo.mallet --num-topics 50 --output-doc-topics /tmp/doc-topics-hugo.txt --output-topic-keys /tmp/topic-keys-hugo.txt --word-topic-counts-file /tmp/word-topic-hugo.txt
```


As described in Chapter ?, We post-process the MALLET output in order to get
a matrix of topic proportions for each document and a matrix of topic-word
associations for each topic.  Each row of the document-topic matrix holds the
topic proportions associated with a document. (Here each document is a 1,000
word section of *Les Misérables*.) Each row of the topic-word matrix holds the
probabilities describing the association of each word in the vocabulary with
a topic. (Recall that the vocabulary is the set of words occurring one or more
times in the corpus.)


````python
import pandas as pd
doc_topic = pd.read_csv('data/les-miserables-doc-topic.csv', index_col=0)
topic_word = pd.read_csv('data/les-miserables-topic-word.csv', index_col=0)
````

Among the fifty topics there is one topic (#35 using 0-based indexing) that jumps out as characteristic of events towards the close of the novel. The words most strongly connected with this topic include “barricade”, “fusil”, and “cartouches” (“barricade”, “rifle”, and “cartridges”).

````python
word_topic35 = topic_word.iloc[35]
# topic35_words is a Series rather than a DataFrame
word_topic35.sort(ascending=False)
word_topic35.head(15)
````

Because the documents are ordered in a sequence, we can plot the fate, so to speak, of this topic over time with the following lines of code:

````python
series = doc_topic.iloc[:, 35]
````

````python
series.plot(use_index=False)
````

While this visualization communicates the essential information about the prevalence of a topic in the corpus, it is not perfect. We can improve it. It would, for instance, be useful to include an indication of where the various volumes start and end. Another enhancement would add some kind of “smoothing” to the time series in order to better communicate the underlying trend.

A rolling average of the topic shares turns out be a useful form of smoothing in this case. We are interested in the prevalence of a topic over time and whether a topic disappears completely in one 500-word chunk of text (only to reappear in the next) does not interest us. We want to visualize the underlying trend, that is, we need some model or heuristic capable of capturing the idea that the topic (or any similar feature) has an underlying propensity to appear at varying points of the novel and that while this propensity may change over time it does not fluctuate wildly. [2]

Recall that a rolling or moving average of a time series associates with each point in the series the average of some fixed number of previous observations (including the current observation). This fixed number of observations is often called a “window”. The idea of a rolling mean (conveniently implemented in pandas.rolling_mean()) is effectively communicated visually:

````python
z = np.array([  3.,   2.,   3.,   6.,   2.,   3.,   1.,   3.,   8.,   3.,   5.,
                8.,   7.,   8.,   7.,   6.,   8.,   7.,   7.,   5.,   8.,   6.,
                11.,   6.,   7.,   8.,   8.,   6.,   9.,  15.,  13.,  10.,   9.]) 
````

````python
pd.rolling_mean(z, 3)
````

````python
plt.plot(z, '.', alpha=0.5)
plt.plot(pd.rolling_mean(z, 5), '-', linewidth=2)
````

After making these two improvements—marking the volume boundaries and adding a trend line based on a rolling average—the time series for our topic does a better job of orienting us in the novel and communicating the points in the novel where the topic appears:

````python
xs = np.arange(len(series))
series_smooth = pd.rolling_mean(series, 15)  # 15 seems to work well here
volume_names = ["tome-1-fantine", "tome-2-cosette", "tome-3-marius", "tome-4", "tome-5-jean-valjean"]
volume_indexes = []
for volname in volume_names:
    for i, docname in enumerate(doc_topic.index):
        if volname in docname:
            volume_indexes.append(i)
            # after we've hit the first document from the volume, go onto the next
            break
````

````python
# method #2, use NumPy functions
for volname in volume_names:
    volume_indexes.append(np.min(np.nonzero([volname in name for name in doc_topic.index])))
````

````python
plt.plot(series, '.', alpha=0.3)
plt.plot(series_smooth, '-', linewidth=2)
plt.vlines(volume_indexes, ymin=0, ymax=np.max(series))
text_xs = np.array(volume_indexes) + np.diff(np.array(volume_indexes + [max(xs)]))/2
text_ys = np.repeat(max(series), len(volume_names)) - 0.05
for x, y, s in zip(text_xs, text_ys, volume_names):
    plt.text(x, y, s, horizontalalignment='center')
plt.title('Les Misérables, Topic #35 (barricade enjolras ...)')
plt.ylabel("Topic share")
plt.xlabel("Novel segment")
# final plot adjustments
plt.ylim(0, max(series));
````
